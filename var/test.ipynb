```python
import numpy as np
import matplotlib as mpl

import matplotlib.pyplot as plt



H0=1
w0=1e3
Q=1

H = lambda s: H0*w0/Q*s/(s**2+w0/Q*s+w0**2)




s=np.linspace(1,1e2,int(1e6))


import cmath
fig = plt.figure()

magax = fig.add_subplot()

magax.semilogx(10*np.log10(np.abs(H(s*1j))),color='k')
magax.set_ylabel('$10\cdot\log_{10}(|H(s)|)$ (Db)')
phax = magax.twinx()
phax.semilogx(np.angle(H(s*1j)),color='r')
phax.set_ylabel('$\phi(s)$ (rad)')
#phax.axhline(y=np.pi/4,linestyle=':',color='r')

10*np.log10(np.abs(H(s)))
```
